## bubble plots

```python
def bubble(x,y,z,df,hue=None,palette=None):
    plt.figure(figsize=[12,6],dpi=300)
    sns.scatterplot(x=x,y=y,size=z,hue=hue,data=df,palette=palette)
    plt.show()
```

```python
bubble('SalePrice','LotArea','LotFrontage',data)
```

```python
bubble('SalePrice','GarageArea','GarageCars',data)
```

```python
bubble('LotArea','GarageArea','OverallQual',data)
```
