```python
import pandas as pd
import numpy as np
import scipy.stats as ss
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_pickle('../data/cleaned_train.pkl')
data.head()
```

```python
data['LotArea']
```

```python
# scott's rule used to divide bins
bins = 3.49 * np.std(data['LotArea']) * (data.shape[0] ** (-1/3))

plt.figure(figsize=[12,6],dpi=300)
sns.histplot(x=data['LotArea'],binwidth=bins,kde=True,stat='probability')
plt.show()
```

```python
# density distribution
plt.figure(figsize=[12,6],dpi=300)
sns.kdeplot(x=data['LotArea'],shade=True)
plt.show()
```
